```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
url = 'https://www.thrillophilia.com/places-to-visit-in-kathmandu'
driver.get(url)
```

```python
#to check if such path is present or not
from selenium.common.exceptions import NoSuchElementException
def hasXpath(xpath):
    try:
        driver.find_element(By.XPATH, xpath)
    except NoSuchElementException:
        return False
    return True
```

```python
# checking if above function works or not
#hasXpath('//*[@id="undefined2"]')
```

```python
# clicking button
def click_button(button_path):
    driver.implicitly_wait(10)
    driver.find_element(By.XPATH,button_path).click()
```

```python
#cclick on the Read More buttons in the website
for x in range(1,55):
    button_path_1 = f'//*[@id="main"]/section[5]/div/div/div/div[{x}]/div[3]/div/a'
    button_path_2 = f'//*[@id="main"]/section[5]/div/div/div/div[{x}]/div[4]/div/a'
    
    # to validate the Xpath of the read more buttons
    if hasXpath(button_path_1):
        true_path = button_path_1
    elif hasXpath(button_path_2):
        true_path = button_path_2
    elif hasXpath(button_path_1+'/span'):
        true_path = button_path_1+'/span'
        
    #calling button click function
    click_button(true_path)
```

```python
# to extract the description, location, timings and pricings
place_list = []
c = 1
for x in range(1,57):
    print(c)
    c += 1
    curr_path_1 = f'//*[@id="main"]/section[5]/div/div/div/div[{x}]/div[3]/div'
    curr_path_2 = f'//*[@id="main"]/section[5]/div/div/div/div[{x}]/div[4]/div'
    #checking if the path exists
    if hasXpath(curr_path_1):
        true_path = curr_path_1
    else:
        true_path = curr_path_2
    
    #now scrapping data of each location
    places = driver.find_element(By.XPATH, true_path)
    print(places.text)
    place_list.append(places.text)
```

```python
c = 0
for x in place_list:
    c += 1
print(c)
```

```python
#remove some of the advertisement that were scrapped from the website
scrapped_data = place_list
for place in scrapped_data:
    if(len(place)<80):
        scrapped_data.remove(place)
```

```python
# to remove the unwanted strings or characters from the scrapped data

from functools import reduce
import re

def remove_char(lst, char):
    return reduce(lambda x, y: x + [re.sub(char, '', y)], lst, [])

#initialize character
char_list = ['Read Less', '\n', ':', '-', 'Highlights']

for char in char_list:
    scrapped_data = remove_char(scrapped_data, char)
```

```python
print(scrapped_data)
```

```python
#splitting the data based on multiple delimiters such as : location, price, timings
splitted_data = [re.split('Location|Timings|Timing|Prices|Price|Entry|Entry Fee', place) for place in scrapped_data]
```

```python
print(splitted_data)
```

```python
import openpyxl
from openpyxl import Workbook

#Create xlxs file
wb_test = Workbook()
wb_test.save('thrillo_description.xlsx')

# Open the Excel file
excel_file_path = 'thrillo_description.xlsx'
workbook = openpyxl.load_workbook(excel_file_path)
sheet = workbook.active

# Write data to Excel
for row_index, row_data in enumerate(splitted_data, start=2):  # Start from row 2 to skip headers
    for col_index, value in enumerate(row_data, start=1):
        sheet.cell(row=row_index, column=col_index, value=value)

# Save the changes
workbook.save(excel_file_path)
```
